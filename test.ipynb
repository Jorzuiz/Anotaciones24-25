```python
import matplotlib.pyplot as plt

def calcular_reduccion_dano(armadura):
    """Calcula la reduccin de dao segn la frmula dada."""
    if armadura < 0:
        return 0  # Evitar divisiones por cero o valores negativos
    return (armadura * 100) / (100 + armadura)

# Generar valores de armadura de 1 a 500
armadura_valores = range(1, 101)
reduccion_valores = [calcular_reduccion_dano(a) for a in armadura_valores]

# Crear la grfica
plt.figure(figsize=(10, 10))  # Ajustar el tamao de la figura
plt.plot(armadura_valores, reduccion_valores, label="Reduccin de Dao", color='blue') #Color azul a la grafica principal

# Aadir la frmula como texto en la grfica
plt.text(250, 60, r'$Reduccin = \frac{Armadura \times 100}{100 + Armadura}$', fontsize=12, color='red', horizontalalignment='center')# Formula en rojo

# Aadir etiquetas y ttulo
plt.xlabel("Armadura")
plt.ylabel("Reduccin de Dao (%)")
plt.title("Relacin entre Armadura y Reduccin de Dao")

# Aadir una leyenda
plt.legend()

# Aadir una cuadrcula para mejor visualizacin
plt.grid(True)

# Mostrar la grfica
plt.show()

#Mostrar valores concretos
print("Ejemplos:")
print("Con 20 de armadura: ", calcular_reduccion_dano(20), "% de reduccion")
print("Con 100 de armadura: ", calcular_reduccion_dano(100), "% de reduccion")
print("Con 300 de armadura: ", calcular_reduccion_dano(300), "% de reduccion")
print("Con 500 de armadura: ", calcular_reduccion_dano(500), "% de reduccion")
```
