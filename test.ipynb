```python
import matplotlib.pyplot as plt

def calcular_reduccion_dano(armadura):
    """Calcula la reducción de daño según la fórmula dada."""
    if armadura < 0:
        return 0  # Evitar divisiones por cero o valores negativos
    return (armadura * 100) / (100 + armadura)

# Generar valores de armadura de 1 a 500
armadura_valores = range(1, 101)
reduccion_valores = [calcular_reduccion_dano(a) for a in armadura_valores]

# Crear la gráfica
plt.figure(figsize=(10, 10))  # Ajustar el tamaño de la figura
plt.plot(armadura_valores, reduccion_valores, label="Reducción de Daño", color='blue') #Color azul a la grafica principal

# Añadir la fórmula como texto en la gráfica
plt.text(250, 60, r'$Reducción = \frac{Armadura \times 100}{100 + Armadura}$', fontsize=12, color='red', horizontalalignment='center')# Formula en rojo

# Añadir etiquetas y título
plt.xlabel("Armadura")
plt.ylabel("Reducción de Daño (%)")
plt.title("Relación entre Armadura y Reducción de Daño")

# Añadir una leyenda
plt.legend()

# Añadir una cuadrícula para mejor visualización
plt.grid(True)

# Mostrar la gráfica
plt.show()

#Mostrar valores concretos
print("Ejemplos:")
print("Con 20 de armadura: ", calcular_reduccion_dano(20), "% de reduccion")
print("Con 100 de armadura: ", calcular_reduccion_dano(100), "% de reduccion")
print("Con 300 de armadura: ", calcular_reduccion_dano(300), "% de reduccion")
print("Con 500 de armadura: ", calcular_reduccion_dano(500), "% de reduccion")
```
